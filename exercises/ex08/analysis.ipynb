# Define a mask definition filtering people rate very low on effectiveness on the assignment.
People who responded 1 or 2 will be considered here. If they did, True will be appended to the list. And the list will be as the same length of the response list.

```python
def filter_row_low(row: list[str]) -> list[bool]:
    """This will go through the respondents who responded 1 and 2 in the survey on prgramming_effective and produce a list of booleans."""
    result: list[bool] = []
    for item in row:
        if item == "1" or item == "2":
            result.append(True)
        else:
            result.append(False)
    return result
```

# Apply the mask definition to the table i want to use.

```python
low_eff = filter_row_low(headed["programming_effective"])
low_eff
```

# Make another definition that dooes the filtering with the mask function.

```python
def filter_table_low(og_table: dict[str, list[str]], mask: list[bool]) -> dict[str, list[str]]:
    """Now return the valubilities of people who think the assignments are not effective."""
    result: dict[str, list[str]] = {}
    for key in og_table:
        result[key] = []
        i = 0
        while i < len(og_table[key]):
            if mask[i] is True:
                result[key].append(og_table[key][i])
            i += 1
    return result
```

# Make a dictionary and see the valubilities only responded by people rated "1" or "2" on the assignment effectiveness.

```python
table_low_eff = filter_table_low(headed,low_eff)
```

# Count how many poeple responded for each criteria.

```python
value_counts: dict[str, int] = count(table_low_eff["valuable"])
print(value_counts)
```
